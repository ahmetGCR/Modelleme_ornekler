## Polynomial regression

```python
import pandas as pd
import numpy as np
```

```python
# veriyi oku
df = pd.read_csv('Power_Plant.csv', delimiter=",")
df.head()
```

```python
# girdileri ve çıktı sutunlarını x ve y değişkenine ata
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
# Veri setini Eğitim seti ve Test seti olarak ayırma
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Standard Scaling

from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
x_train_scaled = standard_scaler.fit_transform(x_train)
x_test_scaled = standard_scaler.transform(x_test)
```

```python
# İlgili kütüphaneyi import et.
from sklearn.preprocessing import PolynomialFeatures
poly_reg_feature_builder = PolynomialFeatures(degree = 5)
x_train_poly = poly_reg_feature_builder.fit_transform(x_train_scaled)
x_train_scaled
```

```python
# Polinom Regresyon modelini eğitin

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train_poly, y_train)
```

```python
# Test datasına ait polinom özelliklerini hazırla ve Test verilerini kullanarak çıktıyı tahmin et

x_test_poly = poly_reg_feature_builder.fit_transform(x_test_scaled)

y_pred = regressor.predict(x_test_poly)
```

```python
y_pred_df = pd.DataFrame(data=y_pred,columns=["Tahmini_Sonuc"])
```

```python
y_test_df = pd.DataFrame(data=y_test,columns=["Gerçek_Sonuç"])

karsılastırma_df = pd.concat([y_test_df,y_pred_df],axis=1)
karsılastırma_df
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
